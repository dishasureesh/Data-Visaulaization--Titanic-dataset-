```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
titanic =sns.load_dataset('titanic')
```

```python
titanic.columns
```

```python
titanic.head()
```

```python
axis = plt.subplots(figsize=(11,5))
sns.boxplot(x='survived',y='age',data=titanic)
axis=plt.subplots(figsize =(10,8))
sns.barplot(x='sex',y='survived',data = titanic)
```

```python
survived = titanic.loc[titanic['survived']==1,'age'].dropna()
```

```python
#Distribution
sns.distplot(survived)
```

```python
survived.skew(),survived.kurt()
```

```python
sns.pairplot(titanic.dropna())
```

```python
#pclass vs survived
grid = sns.FacetGrid(titanic,col='survived',row='pclass',aspect=1.5)
grid.map(plt.hist,'age',alpha=0.8,bins=20)
grid.add_legend()
```
